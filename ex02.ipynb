```python
import pandas as pd

file = pd.read_excel("empresa.xlsx")
display(file)
```

2- Drawing a pizza graph with the expenses of July/2024

```python
import matplotlib.pyplot as plt

sector = file["Setor"]
expense = file["Gasto"]

#Drawing the graph

plt.title("Monthly Expenses by Sector - July/2024")
plt.pie(expense, labels=sector, autopct="%1.2f%%")
plt.show()
```
