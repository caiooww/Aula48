1 - Opening an Excel file with Pandas.

```python
import pandas as pd

file = pd.read_excel("dolar.xlsx")
display(file)
```

2- Generating a graph with dolar buy value of 2023/July

```python
import matplotlib.pyplot as plt

dolarBuy = file["Compra"]

#Drawing the graph

plt.title("Dolar Buy Value - July/2023")
plt.plot(dolarBuy, marker='o', linestyle='dashed', color='purple')
plt.grid()
plt.show()
```

3- Generating a graph with dolar sell value of July/2023

```python
dolarSell = file["Venda"]

#Drawing the graph

plt.title("Dolar Sell Value - July/2023")
plt.plot(dolarSell, marker='o', linestyle='dashed', color='red')
plt.grid()
plt.show()
```
