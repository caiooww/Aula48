1- Opening an excel file

```python
import pandas as pd

table1 = pd.read_excel("vendas.xlsx")
display(table1)
```

2- calculating total store revenue

```python
import matplotlib.pyplot as plt

totalRevenue = table1["Valor Total"].sum()
print(f"Total Revenue: {totalRevenue:.2f}")
```

3- Calculating the revenue of a store

```python
byStore = table1[["Loja", "Valor Total"]].groupby("Loja").sum()
display(byStore)
```

4- Revenue of each product by store

```python
byStore = table1[["Loja", "Produto", "Quantidade", "Valor Total"]].groupby(["Loja", "Produto"]).sum()
display(byStore)

#Generating a sheet
byStore.to_excel("sellsPerStore.xlsx")
```
